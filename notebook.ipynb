```python
%pip install numpy

import numpy as np

scores = np.array([
    [80, 85, 90],
    [70, 60, 75],
    [95, 90, 100],
    [40, 50, 45]
])

avg = np.average(scores, axis=1)
print(f"Nilai Rata Rata: {avg}")

max = np.max(scores)
print(f"Nilai maksimal: {max}")

filter = np.where(avg >= 70, "Lulus", "Gagal")
print(f"Data setelah di seleksi: {filter}")
```

```python
%pip install numpy pandas

import pandas as pd
import numpy as np

df_data = {
    'Nama': ['Andi', 'Budi', 'Caca', 'Dedi', 'Euis'],
    'Departemen': ['IT', 'HR', 'IT', 'Sales', 'Sales'],
    'Gaji': [8000000, 7000000, 8500000, np.nan, 6000000],
    'Pengalaman_Tahun': [5, 3, 6, 2, 1]
}

df = pd.DataFrame(df_data)

df['Gaji'] = df['Gaji'].fillna(df['Gaji'].mean())
print(df)

df['Kategori_Senoritas'] = np.where(df['Pengalaman_Tahun'] > 4, 'Senior', 'Junior')
print(df)

df_avg_salary = df.groupby('Departemen') ['Gaji'].mean()
print(df_avg_salary)
```
